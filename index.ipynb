# Stack Overflow Developer Survey 2020

This is my attempt to gain insights from Stack Overflow Developer Survey in 2020. They have been doing this since 2011. However, current version won't go that far. In fact, my main focus is to compare the trends in Indonesia (a country where I was born and raised) with what's happening around the world.

```python
# import module
import src.stackoverflow as ss
```

```python
# assign object classes
data = ss.Dataset()
viz = ss.Visualization()

# load dataframe
world = data.load_data(is_indonesia=False)
indonesia = data.load_data(is_indonesia=True)
```

```python
# create dataframe from gender column
gender_world = data.group_by(world, 'Gender', 'gender')
gender_indonesia = data.group_by(indonesia, 'Gender', 'gender')

# make a nice visualization thx to xkcd
viz.bar_chart(gender_world, gender_indonesia, 'gender', caption_x=-0.5, caption_y=-0.45)
```

# Age First Code

It looks like that people started coding in their teenage years. However, it's interesting that a few people have the willingness to learn even though they're not young anymore. What drives them to pick up programming in their 60s to 80s? How can this be replicated to Indonesians?

```python
data.age_first_code(world).describe().to_frame().transpose()
```

```python
data.age_first_code(indonesia).describe().to_frame().transpose()
```

```python
age_world = data.age_first_code(world)
age_indonesia = data.age_first_code(indonesia)
viz.visualize_age(
    age_world=age_world,
    age_indonesia=age_indonesia,
    label_world='World',
    label_indonesia='Indonesia',
    suptitle='Age distribution when started to code',
    caption='Vertical line shows median age'
    
)
```

# Education Level

It makes sense that majority of the respondents have bachelor degree in the tech world. It's even more interesting to see more Indonesian developers never completed any formal education than the rest of the world. I don't know whether the barrier to entry is lower here than anywhere in this world or not, though. But it makes sense to some extent considering higher education is still a luxury in Indonesia.

```python
# create dataframe from gender column
educ_world = data.group_by(world, 'EdLevel', 'education')
educ_indonesia = data.group_by(indonesia, 'EdLevel', 'education')

# make a nice visualization thx to xkcd
viz.bar_chart(educ_world, educ_indonesia, 'education', caption_x=-0.5, caption_y=-0.325)
```
