# Stack Overflow Developer Survey 2020

This is my attempt to gain insights from 2020 Stack Overflow Developer Survey in 2020. They have been doing this since 2011. However, current version won't go that far. In fact, my main focus is to compare the trends in Indonesia (a country where I was born and raised) with what's happening around the world.

```python
# import module
import src.stackoverflow as ss
```

```python
# assign object classes
data = ss.Dataset()
viz = ss.Visualization()

# load dataframe
world = data.load_data(is_indonesia=False)
indonesia = data.load_data(is_indonesia=True)
```

# Gender

```python
# create dataframe from gender column
gender_world = data.gender_dataframe(world)
gender_indonesia = data.gender_dataframe(indonesia)

# make a nice visualization thx to xkcd
viz.visualize_gender(gender_world, gender_indonesia)
```

# Age First Code

```python
data.age_first_code(world).describe().to_frame().transpose()
```

```python
data.age_first_code(indonesia).describe().to_frame().transpose()
```

```python
age_world = data.age_first_code(world)
age_indonesia = data.age_first_code(indonesia)
viz.visualize_age(
    age_world=age_world,
    age_indonesia=age_indonesia,
    label_world='Selain Indonesia',
    label_indonesia='Indonesia',
    
)
```
