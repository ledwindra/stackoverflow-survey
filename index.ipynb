```python
# import module
import src.stackoverflow as ss

# initialize object classes
data = ss.Dataset()
viz = ss.Visualization()
```

```python
# load dataframe
world = data.load_data(is_indonesia=False)
indonesia = data.load_data(is_indonesia=True)
```

# Gender

What the hell? We have big names such as [Ada Lovelace](https://en.wikipedia.org/wiki/Ada_Lovelace) and [Margareth Hamilton](https://en.wikipedia.org/wiki/Margaret_Hamilton_(software_engineer)) as pioneers in computer science and software engineering and this is what we get? Although there has got to be a better methodology to make a survey, I think this one captures the reality in the field and industry.

Now let's look carefully. Majority of the respondents in both Indonesia and the rest of the world are men. Quite many of them choose not to answer their gender, which I hunch they might think it's a sensitive information to be collected for such survey or irrelevant to judge whether they are good developers or not. Only a small fraction of them self-identify as a woman; and non-existent non-binary, genderqueer, or gender non-conforming hardly appear on the graph among both Indonesians and the people around the world.

I self-identify as a man, and of course I don't have any expertise in the field of gender studies or so. It's my attempt to educate myself and learn to do better. Would be happy if the readers here would share something about the issue üôèüèΩ.

```python
# a set of values in terms of gender
set(world['Gender'].to_list() + indonesia['Gender'].to_list())
```

```python
# create dataframe from gender column
gender_world = data.transform(
    data.group_by(world, 'Gender', 'gender'),
    is_in=['Man', 'Woman', 'N/A'],
    column='gender',
    value='Non-binary, genderqueer, or gender non-conforming'
)
gender_indonesia = data.transform(
    data.group_by(indonesia, 'Gender', 'gender'),
    is_in=['Man', 'Woman', 'N/A'],
    column='gender',
    value='Non-binary, genderqueer, or gender non-conforming'
)

# make a nice visualization thx to xkcd
viz.bar_chart(gender_world, gender_indonesia, 'gender', 'Gender', caption_x=-0.25, caption_y=-0.3)
```

# Sexuality
I have a hunch here that many people choose not to answer because it's a sensitive information. Also this is an interesting issue to investigate further. Generally speaking, being openly not straight in Indonesia can still have negative consequences.
```python
# a set of values in terms of sexuality
set(world['Sexuality'].to_list() + indonesia['Sexuality'].to_list())
```

```python
# create dataframe from sexuality column
sexuality_world = data.transform(
    data.group_by(world, 'Sexuality', 'sexuality'),
    column='sexuality',
    is_in=['Straight / Heterosexual', 'N/A'],
    value='Lesbian, Gay, Bisexual, or Queer'
)
sexuality_indonesia = data.transform(
    data.group_by(indonesia, 'Sexuality', 'sexuality'),
    column='sexuality',
    is_in=['Straight / Heterosexual', 'N/A'],
    value='Lesbian, Gay, Bisexual, or Queer'
)
viz.bar_chart(sexuality_world, sexuality_indonesia, 'sexuality', 'Sexuality', caption_x=-0.25, caption_y=-0.2)
```

# Education Level

It makes sense that majority of the respondents have bachelor degree in the tech world. It's even more interesting to see more Indonesian developers never went to university more than the rest of the world. I don't know whether the barrier to entry is lower here than anywhere in this world or not, though. But it makes sense to some extent considering higher education is still a luxury in Indonesia.

```python
# a set of values in terms of education
set(world['EdLevel'].to_list() + indonesia['EdLevel'].to_list())
```

```python
# create dataframe from education column
educ_world = data.transform(
    data.group_by(world, 'EdLevel', 'education'),
    column='education',
    is_in=[
        'Associate degree (A.A., A.S., etc.)',
        'Bachelor‚Äôs degree (B.A., B.S., B.Eng., etc.)',
        'Master‚Äôs degree (M.A., M.S., M.Eng., MBA, etc.)',
        'Other doctoral degree (Ph.D., Ed.D., etc.)',
        'Professional degree (JD, MD, etc.)',
        'Some college/university study without earning a degree',
        'N/A'
    ],
    value='Never went to university'
)
educ_indonesia = data.transform(
    data.group_by(indonesia, 'EdLevel', 'education'),
    column='education',
    is_in=[
        'Associate degree (A.A., A.S., etc.)',
        'Bachelor‚Äôs degree (B.A., B.S., B.Eng., etc.)',
        'Master‚Äôs degree (M.A., M.S., M.Eng., MBA, etc.)',
        'Other doctoral degree (Ph.D., Ed.D., etc.)',
        'Professional degree (JD, MD, etc.)',
        'Some college/university study without earning a degree',
        'N/A'
    ],
    value='Never went to university'
)

# make a nice visualization thx to xkcd
viz.bar_chart(educ_world, educ_indonesia, 'education', 'Education', caption_x=-0.5, caption_y=-0.325)
```
